1 : What is Simple Linear Regression (SLR)? Explain its purpose.


Simple Linear Regression (SLR) :

Simple Linear Regression (SLR) is a statistical method used to model and analyze the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the observed data.

Mathematical Model

The relationship is represented by the equation:
                                                 Y=Œ≤0+Œ≤1X+Œµ

Where:

Y = Dependent (response) variable
X = Independent (predictor) variable
Œ≤0 = Intercept (value of Y when X = 0)
ùõΩ1= Slope (change in Y for a one-unit change in X)
Œµ = Error term (random variation)

Purpose of Simple Linear Regression

The main purposes of SLR are:

To understand the relationship
It helps determine whether and how strongly the independent variable affects the dependent variable.

To make predictions
Once the relationship is established, SLR can be used to predict the value of ùëå for a given value of ùëã

To quantify the effect
The slope (ùõΩ1) shows how much the dependent variable changes when the independent variable increases by one unit.

To explain variation
It explains how much of the variation in the dependent variable is explained by the independent variable.

Example
If:

X = Hours studied
Y = Exam score

SLR helps estimate how exam scores change as study hours increase.

2: What are the key assumptions of Simple Linear Regression?

Key Assumptions of Simple Linear Regression (SLR)

Simple Linear Regression relies on several important assumptions to ensure that the model is valid and reliable.

- Linearity :

The relationship between the independent variable (X) and the dependent variable (Y) is linear.
This means changes in ùëåare proportional to changes in ùëã

 Checked using: Scatter plot or residuals vs. X plot.

- Independence of Errors :

The residuals (errors) are independent of each other.
One observation‚Äôs error should not influence another‚Äôs.

 Common issue: Autocorrelation in time-series data.
 Checked using: Durbin‚ÄìWatson test.

- Homoscedasticity:

The variance of residuals is constant across all levels of ùëã
Errors should not fan out or funnel in.

 Checked using: Residuals vs. fitted values plot.

- Normality of Errors :

The residuals are normally distributed.
This assumption is especially important for hypothesis testing and confidence intervals.

 Checked using: Histogram, Q‚ÄìQ plot, or Shapiro‚ÄìWilk test.

- No Multicollinearity:

Since SLR has only one independent variable, multicollinearity is not applicable.
(This becomes relevant in multiple linear regression.)

- Zero Mean of Errors:

The average value of residuals is zero:

E(Œµ)=0

Mathematical Equation of Simple Linear Regression (SLR)

The simple linear regression model is given by: Y=Œ≤0+Œ≤1X+Œµ

Explanation of Each Term

Y ‚Äî Dependent Variable (Response)

Œ≤0 ‚Äî Intercept
The expected value of Y when ùëã=0.

Œ≤1‚Äî Slope (Regression Coefficient)

Œµ ‚Äî Error Term (Residual)
Accounts for the random variation in ùëå that cannot be explained by ùëã
It includes measurement errors, omitted variables, and noise.

After fitting the model to data, the equation becomes: Y^=Œ≤^0+Œ≤^1X

6: What is Logistic Regression? How does it differ from Linear Regression?

Logistic Regression

Logistic Regression is a statistical and machine learning technique used for binary classification problems, where the dependent variable has two possible outcomes (e.g., Yes/No, 0/1, Pass/Fail).

Instead of predicting a continuous value, logistic regression predicts the probability that an observation belongs to a particular class.

Example :

Linear Regression: Predicting house price based on area
Logistic Regression: Predicting whether a customer will buy a product (Yes/No)

Why Not Use Linear Regression for Classification?

Linear regression can predict values outside [0,1]
It violates assumptions when applied to categorical outcomes
Logistic regression ensures valid probabilities

8: What is the purpose of the R-squared metric in regression analysis?


Purpose of the R-squared Metric in Regression Analysis :

R-squared is a statistical measure that indicates how well a regression model explains the variability of the dependent variable using the independent variable(s).

What R-squared Measures :
It represents the proportion of the total variation in the dependent variable (Y) that is explained by the regression model.

Where:

SSres = Sum of squared residuals (unexplained variation)

SStot = Total sum of squares (total variation)

Purpose of R-squared

Model Goodness of Fit
Shows how well the model fits the observed data.

Explained Variance
Indicates the percentage of variation in ùëå explained by 


Model Comparison
Helps compare different regression models on the same dataset (higher R2 usually indicates better fit).

- Interpretability
Easy to understand as a percentage (e.g.,ùëÖ2 = 0.85 means 85% of variance is explained).

Interpretation Example

R2= 0.90: Very strong explanatory power
ùëÖ2 = 0.50: Moderate explanatory power
R2 = 0.10: Weak explanatory power

Limitations of R-squared

A high ùëÖ2 does not guarantee a good model
It does not indicate causation
It always increases when more predictors are added (even irrelevant ones)

9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (X = independent variable, y = dependent variable)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # reshape for sklearn
y = np.array([2, 4, 6, 8, 10])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

10: How do you interpret the coefficients in a simple linear regression model?

Interpreting the Coefficients in a Simple Linear Regression (SLR) Model

A simple linear regression model is written as:  Y=Œ≤0+Œ≤1X+Œµ

Each coefficient has a clear and practical meaning.

Intercept (ùõΩ0)

The intercept represents the expected value of the dependent variable ùëå when the independent variable X=0.
It shows where the regression line crosses the Y-axis.

Example:
If ùõΩ0 =40, then when 
X=0, the predicted value of Y is 40.

 Note: Sometimes 
X=0 may not be meaningful in real life (e.g., 0 years of experience).

Slope (Œ≤1)

The slope indicates the average change in Y for a one-unit increase in ùëã
It tells both the direction and strength of the relationship.

Interpretation:
ùõΩ1<0: Negative relationship (Y decreases as X increases)
ùõΩ1=0: No linear relationship

Example:
If ùõΩ1=5, then for every 1-unit increase in ùëã, Y increases by 5 units.
